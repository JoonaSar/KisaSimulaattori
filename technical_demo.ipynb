## Kisan simulointi

Simuloidaan kisaa jonona eri vartio-joukoille määriteltyinä kisaosuuksina, eli tehtävinä ja suunnistuksina. Ne toimivat molemmat hyvin samanlaisesti: vartiot tulevat jollain aikajakaumalla sisään, jonka jälkeen tehtävä tai suunnistusosuus simuloi vartion viettämää aikaa osuudella. Tämä aika lisätään saapumisaikaan, jonka tulosta käytetään taas seuraavan rataosuuden saapumisaikana. Jokainen kisan osuus vartioiden omia taukoja lukuunottamatta voidaan simuloida tällä tavoin.


### Esimerkki 1
Tehtävä, joka on määritelty ainoastaan punaiselle sarjalle. Kuvaajassa laskennalliset sisääntuloajat ennen ja jälkeen tehtävän. 

```python
import pandas as pd
import numpy as np 
import seaborn as sns
from numpy.random import default_rng
import matplotlib.pyplot as plt
```

```python
vartiot = pd.DataFrame(columns=["sarja", "aika"])
rng = default_rng(0)
vartiot["aika"] = rng.integers(low=0, high=60, size=100)
vartiot["sarja"] = rng.choice(["Punainen", "Sininen"], size=100)


f, ax = plt.subplots(2,2, figsize=(13, 13))

#sns.scatterplot(x=vartiot["aika"], y=vartiot["aika"], hue = vartiot["sarja"], ax = ax[0])
sns.violinplot(data=vartiot, y="sarja", hue="sarja", x="aika", orient="h", ax=ax[0][0])
sns.histplot(data=vartiot, x="aika", hue="sarja", multiple="stack", ax=ax[1][0])

mask = vartiot["sarja"]=="Punainen"
vartiot.loc[mask, "aika"] += rng.integers(low=20, high=40, size=mask.sum())
vartiot


#sns.scatterplot(x=vartiot["aika"], y=, hue = vartiot["sarja"], ax = ax[1])
sns.violinplot(data=vartiot, y="sarja", hue="sarja", x="aika", orient="h", ax=ax[0][1])
sns.histplot(data=vartiot, x="aika", hue="sarja", multiple="stack", ax=ax[1][1])
```
